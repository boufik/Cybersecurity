# DataFrame

```python
df = pd.DataFrame(allowed_triples, columns=["source_ref", "relationship_type", "target_ref"])
print(df)
```

# Functions

```python
def print_filtered(sentences, filter_by):
  print(f"Filtering by:  {filter_by}\n")
  for counter, sentence in enumerate(sentences):
    print(f"{counter + 1}) {sentence}")
```

## 1) source_ref

```python
def get_source_ref_instances(source_ref):
  filter_by = f"source_ref = {source_ref}"
  filtered_df = df[df["source_ref"] == source_ref]
  sentences = list()
  for index, row in filtered_df.iterrows():
    # The variable row is instance of: pandas.core.series.Series
    source_ref = row['source_ref']
    relationship_type = row['relationship_type']
    target_ref = row['target_ref']
    sentence = f"{source_ref} {relationship_type} {target_ref}"
    sentences.append(sentence)
  return sentences, filter_by

sentences, filter_by = get_source_ref_instances("tool")
print_filtered(sentences, filter_by)
```

## 2) relationship_type

```python
def get_relationship_type_instances(relationship_type):
  filter_by = f"relationship_type = {relationship_type}"
  filtered_df = df[df["relationship_type"] == relationship_type]
  sentences = list()
  for index, row in filtered_df.iterrows():
    # The variable row is instance of: pandas.core.series.Series
    source_ref = row['source_ref']
    relationship_type = row['relationship_type']
    target_ref = row['target_ref']
    sentence = f"{source_ref} {relationship_type} {target_ref}"
    sentences.append(sentence)
  return sentences, filter_by

sentences, filter_by = get_relationship_type_instances("targets")
print_filtered(sentences, filter_by)
```

## 3) target_ref

```python
def get_target_ref_instances(target_ref):
  filter_by = f"target_ref = {target_ref}"
  filtered_df = df[df["target_ref"] == target_ref]
  sentences = list()
  for index, row in filtered_df.iterrows():
    # The variable row is instance of: pandas.core.series.Series
    source_ref = row['source_ref']
    relationship_type = row['relationship_type']
    target_ref = row['target_ref']
    sentence = f"{source_ref} {relationship_type} {target_ref}"
    sentences.append(sentence)
  return sentences, filter_by

sentences, filter_by = get_target_ref_instances("location")
print_filtered(sentences, filter_by)
```

## 4) source_ref + target_ref

```python
def get_source_target_ref_instances(source_ref, target_ref):
  filter_by = f"source_ref = {source_ref} && target_ref = {target_ref}"
  filtered_df = df[
    (df["source_ref"] == source_ref) &
    (df["target_ref"] == target_ref)
  ]
  sentences = list()
  for index, row in filtered_df.iterrows():
    # The variable row is instance of: pandas.core.series.Series
    source_ref = row['source_ref']
    relationship_type = row['relationship_type']
    target_ref = row['target_ref']
    sentence = f"{source_ref} {relationship_type} {target_ref}"
    sentences.append(sentence)
  return sentences, filter_by

sentences, filter_by = get_source_target_ref_instances("malware", "tool")
print_filtered(sentences, filter_by)
```

```python
def play():
  dashes = 80*'-'
  first_line = "**** STIX 2.1 relations between SDOs - valid SROs relationship_type ****"
  press = f"Press (...) to filter by (...):"
  options = f"1. source_ref\n2. relationship_type\n3. target_ref\n4. source_ref + target_ref"
  valid_options = ["1", "2", "3", "4", "stop", "exit", "close", ""]
  while True:
    print(f"{dashes}\n{first_line}\n{press}\n\n{options}\n\n")
    choice = input(">> Select (1, 2, 3, 4): ")
    if choice not in valid_options:
      print("Invalid option. Try again.\n")
      continue
    if choice == "1":
      source_ref = input(">> Enter source_ref: ")
      sentences, filter_by = get_source_ref_instances(source_ref)
      print_filtered(sentences, filter_by)
    elif choice == "2":
      relationship_type = input(">> Enter relationship_type: ")
      sentences, filter_by = get_relationship_type_instances(relationship_type)
      print_filtered(sentences, filter_by)
    elif choice == "3":
      target_ref = input(">> Enter target_ref: ")
      sentences, filter_by = get_target_ref_instances(target_ref)
      print_filtered(sentences, filter_by)
    elif choice == "4":
      source_ref = input(">> Enter source_ref: ")
      target_ref = input(">> Enter target_ref: ")
      sentences, filter_by = get_source_target_ref_instances(source_ref, target_ref)
      print_filtered(sentences, filter_by)
    else:
      print("Exiting...")
      break
    print("\n")

play()
```
